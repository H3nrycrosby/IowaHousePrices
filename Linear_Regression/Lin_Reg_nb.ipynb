### Import sklearn libraries

```python
from sklearn.linear_model import ElasticNet, Lasso, LassoCV, Ridge, BayesianRidge, LassoLarsIC
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
```

#### Cross Validation Function

```python
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train_x.values)
    rmse= np.sqrt(-cross_val_score(model, train_x.values, train_y, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

## Regressions

#### Lasso (L1 regularizer)

```python
lasso = make_pipeline(Lasso(alpha =0.0005, random_state=1))
```

```python
score = rmsle_cv(lasso)
print("\nLasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
model_lasso = LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(train_x.values, train_y)
```

```python
coef = pd.Series(model_lasso.coef_, index = train_x.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = pd.concat([coef.sort_values().head(15),
                     coef.sort_values().tail(15)])
```

```python
imp_coef.plot(kind = "barh", figsize = (15,10))
plt.title("Coefficients in the Lasso Model")
```
